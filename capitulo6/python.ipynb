```python
class Foo(object):
    def _get_bar(self):
        return self.__bar
    def _set_bar(self, value):
        if not isinstance(value, int):
            raise TypeError("variável bar deve ser do tipo inteira")
        self.__bar = value
    bar = property(_get_bar, _set_bar)
    
f = Foo()
f.bar = 3
print(f.bar)
f.bar = "three"
```

```python
class Animal:
    def caminha(self):  
        print("caminhando")
        
class Gato(Animal): 
    def pula(self): 
       print("pulando")
    
gato = Gato()# criando instância de Gato
animal = Animal() # criando instância de Animal
gato.caminha() # chamado da classe Animal 
gato.pula() # chamado da classe Gato 
animal.pula() # ERRO, animal não tem o método pula 
```

```python
class Gato(Animal): 
    def __init__(self, cor, olhos):
       self.cor = cor
       self.olhos = olhos
    def pula(self):  
       print(f'pulando {self.cor}')

gato_preto = Gato("preto", "azuis") # criando uma instância de Gato
gato_preto.pula()
gato_branco = Gato("branco", "verde") # criando outra instância de Gato
gato_branco.pula()
```

```python
class Conta(): 
    def __init__(self, ini): 
        self.numero = ini
    def __add__(self, num):
        return Conta(self.numero + num)
    def __str__(self):
        return str(f"str {self.numero}" )
    
conta = Conta(5)
print(conta)
conta += 5
print(conta)
print(conta + 5)
```
