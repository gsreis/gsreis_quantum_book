```python
class Conta(): 
    def __init__(self, ini): 
        self.numero = ini
    def __mul__(self, num):
        return Conta(self.numero * num)
    def __add__(self, num):
        return Conta(self.numero + num)
    def __str__(self):
        return f"{self.numero}"  
    def __contains__(self, num): 
        if self.numero != num: 
            return False
        else: 
            return True
conta = Conta(5)
print(conta * 5)   #chamando método __mul__
print(5 in conta)  #chamando método __contains__ 
```

```python
class Animal:
    def caminha(self):  
        print("caminhando")
        print("caminhando pai")
class Gato(Animal): 
    def __init__(self, cor, olhos):
       self.cor = cor
       self.olhos = olhos
    def pula(self):  
       print(f'pulando {self.cor}')
       super.caminha()
    def caminha(self):  
        super().caminha()
        print("caminhando filho")

gato_preto = Gato("preto", "azuis") # criando uma instância de Gato
gato_branco = Gato("branco", "verde") # criando outra instância de Gato
gato_branco.caminha()
```
